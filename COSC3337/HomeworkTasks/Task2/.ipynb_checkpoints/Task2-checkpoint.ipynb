```python
# Task 2.2: Creating dataset2 of 1000 random samples without replacement
def task2(data, q):
    newData = data.sample(n=q, replace=False)
    newData = newData.reset_index()
    newData = newData.drop('index', axis =1)
    return newData
def task2split(data, q):
    split1, split2 = train_test_split(data, test_size = q)
    return split1, split2
dataset2 = task2(df, 1000)
print("Number of samples in Label 0 in dataset2:", len(dataset2[dataset2['Label'] == 0]))
print("Number of samples in Label 1 in dataset2:", len(dataset2[dataset2['Label'] == 1]))
print("Ratio of Label 0 to Label 1 in dataset2:", round(len(dataset2[dataset2['Label'] == 0])/len(dataset2[dataset2['Label'] == 1]), 2))
```

In task 2.2, I made a two functions that will be valuable later on when we get into creating a function pipeline. The first function task2 takes in a data frame and the size of the sample we want and return a new dataframe of size q that is randomly sampled. I added an extra two lines of code to reset the indexes so that we can reference these cells easier from now on. The second function task2split is used to split the data given a specific test size fraction for q. For example, if I want to split a dataframe for 80% training and 20% testing, I will parse 0.2 as my q value. I then printed out the results the same way I did for the previous task.

```python
# Task 2.3: Stratified sampling
def task3(data, q):
    split = StratifiedShuffleSplit(n_splits = 1, test_size = 1-(q/len(data)))
    for train_index, test_index in split.split(data, data['Label']):
        data0 = data.loc[train_index]
    data0 = data0.reset_index()
    data0 = data0.drop('index', axis =1)
    return data0
dataset3 = task3(df, 1000)
print("Number of samples in Label 0 in dataset2:", len(dataset3[dataset3['Label'] == 0]))
print("Number of samples in Label 1 in dataset2:", len(dataset3[dataset3['Label'] == 1]))
print("Ration of Label 0 to Label 1 in dataset3:", round(len(dataset3[dataset3['Label'] == 0])/len(dataset3[dataset3['Label'] == 1]), 2))
```

In task 2.3, I used a similar function to task 2.3 but imported the function StratifiedShuffleSplit from sklearn.model_selection that automatically splits a dataset in a stratified manner given a test_size. The test size will be a fraction here as well, when I define the split, I use .loc to find the indexes of the stratified sample that I took and create a new dataset data0. I will then reset and drop the index values from the dataset and return the dataset. I then printed out the results the same way I did for the previous two tasks.

```python
# Task 2.4: Reporting covariance matrix of dataset3

covMatrix = dataset3.cov()
sb.heatmap(covMatrix, annot = True, vmax = 8000)
plt.show()
def select2(data, minr = 1, minc = 1):
    min = abs(data["Feature " + str(minr)].cov(data["Feature " + str(minc)]))
    for x in range(4):
        xStr = "Feature " + str(x+1)
        for y in range(4-x):
            yStr = "Feature " + str(y+1)
            if(abs((data[xStr].cov(data[yStr]))) < min):
                min = data[xStr].cov(data[yStr])
                minr = x+1
                minc = y+1
    newData = data[["Feature " + str(minr), "Feature " + str(minc), 'Label']]
    return newData
dataset4 = select2(dataset3)
c = []
for index, row in dataset4.iterrows():
    if row['Label'] == 0:
        c.append(color1)
    else:
        c.append(color2)
dataset4['Color'] = c
```

```python
# Task 2.7: Use the same sampling method used in task 2.3 to extract a test and train set of dataset4
training_set, testing_set = task2split(dataset4, 0.3)
training_set = training_set.drop('Color', axis = 1)
testing_set = testing_set.drop('Color', axis = 1)
```

In task 2.7, I called the function task2split that will stratify split dataset4 with a 30% test split size. I have training_set and testing_set that will take the stratified sampling sets for their respective training and testing set. I then dropped the color attribute from the dataset since we won't be needing that attribute anymore in the upcoming tasks.

```python
# Task 2.8: Train a decision tree with depth = 3 using the training_set dataset and report accuracy using testing_set
def train1(data0, data1):
    colNames = list(data0.columns)
    X = data0.drop('Label', axis = 1)
    y = data0['Label']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
    dtree = DecisionTreeClassifier()
    dtree = dtree.fit(X_train, y_train)
    plt.figure(figsize=(16,12))
    tree.plot_tree(dtree, fontsize = 8, max_depth = 3, feature_names = [colNames[0], colNames[1]])
    plt.show()

    X = data1.drop('Label', axis = 1)
    y = data1['Label']
    testing_set_train0, testing_set_test0, testing_set_train1, testing_set_test1 = train_test_split(X, y, test_size = 0.3)
    y_pred = dtree.predict(testing_set_test0)
    print("Accuracy of training set: ", metrics.accuracy_score(testing_set_test1, y_pred))

def train2(data0, data1):
    colNames = list(data0.columns)
    X = data0.drop('Label', axis = 1)
    y = data0['Label']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
    dtree = DecisionTreeClassifier()
    dtree = dtree.fit(X_train, y_train)
    plt.figure(figsize=(16,12))
    tree.plot_tree(dtree, fontsize = 8, max_depth = 3, feature_names = [colNames[0], colNames[1], colNames[3]])
    plt.show()

    X = data1.drop('Label', axis = 1)
    y = data1['Label']
    testing_set_train0, testing_set_test0, testing_set_train1, testing_set_test1 = train_test_split(X, y, test_size = 0.3)
    y_pred = dtree.predict(testing_set_test0)
    print("Accuracy of training set: ", metrics.accuracy_score(testing_set_test1, y_pred))
train1(training_set, testing_set)

```

In task 2.8, I created two training functions that will be more relevant in future tasks. train1 takes a pair of training and testing sets and using an 80:20 split on the training data and fit a decision tree model by calling the DecisionTreeClassifier() and fitting my X_train and y_train datasets. We will then split the testing dataset the same exacy way and use the model we made with the training sets and apply a .predict() function to the model to predict the testing_set_test0 values. We will then apply the metrics library to find the accuracy score between the testing_set_test 1 and the .predict() value that we had obtained. train1 is relevant for a dataset that hasn't gone through task 2.9 yet since it doesn't accomadate for a new feature. train2 does the same thing but does accomodate for a new feature that we will add in the next task.

```python
# Task 2.9: Write a function that adds a new feature in the dataframe of both training and testing set
def addFeature(data):
    colNames = list(data.columns)
    f = []
    for index, row in data.iterrows():
        f.append(math.sqrt((row[colNames[0]]*row[colNames[0]])+(row[colNames[1]]*row[colNames[1]])))
    data['New Feautre'] = f
    return data
c_training_set = addFeature(training_set)
c_testing_set = addFeature(testing_set)
```

In task 2.9, I made a function that will take a dataframe as a parameter and adds a new column that takes the square root of value1 squarred plus value2 squarred. I did this by starting an empty array and used a for loop to iterate through the rows to calculate each value of the new feature in each row and append to the initially empty array f. I then made a new column 'New Feature' and assigned it to the array f. I called this function for two new datasets, c_training_set and c_testing_set that will add the new feature to the origining training_set and testing_set.

```python
# Task 2.12: write a function that accepts a dataset, "rgl" or "stf", variable for number of samples in the sampled dataset, variable for specifying the number of samples in training set
# and make a pipeline using this function

def task12(data, string, numSamples, numTrain):
    train_num = numTrain/numSamples
    if string == 'rgl':
        dataTrain , dataTest = (data.pipe(task2, numSamples).pipe(select2).pipe(addFeature).pipe(task2split, train_num)) 
        train2(dataTrain, dataTest)
    elif string == 'stf':
        dataTrain, dataTest = (data.pipe(task3, numSamples).pipe(select2).pipe(addFeature).pipe(task2split,train_num))
        train2(dataTrain, dataTest)
    else:
        print("Invalid string input: 'rgl' or 'stf'")
task12(df, "stf", 500, 300)
task12(df, "rgl", 100, 70)
task12(df, "stf", 1500, 1000)
```

In task 2.12, I explored what a function pipeline is. Function pipelines are essentially and series of functions that take an arugment as the previous piping function, in this case, the arguments passed from function to functions is always a dataset of some sort. I made a function called task12 that will take a dataset, a variable specifiying the sampling method of either 'rgl' or 'stf', a variable for number of samples in sampled dataset and a variable specifying the number of samples in training set. In the function, I made an if statement that checks the validity of the string input. If the string is 'rgl', I passed a pipeline that will take a regular random sample (task2) of sample size numSamples and perform a multitude of operations on the dataset identical to what I have been explaining to this moment. if the string is 'stf', I passed a very similar pipeline but instead of the regular random sampling, I used the stratify function task3 to get my sample. We can analyze that for the most part, the regular random sampling model is less accurate than the stratified sampling model no matter how many samples we use for the testing case. Building a pipeline is very efficient way to clean up data from a dataset because you can impose a series of dataset cleanup techniques that will only take one line of code which makes organizaton less cumbersome. 

In conclusion, I have discovered that stratified sampling is more efficient in making a model that is used to predict a class variable in a dataset. Although adding a new feature that computes the sum of the square roots of the two least linearly dependent in the dataset, the stratified random sampling has an advantage over the regular random sampling. I have learned that the covariance is an impoirtant attribute used to find the two variables with the least amount of linear dependency which can be helpful in building a model that is made to predict values based on specific varialbes. When variables are not linearly dependent, the points are more dispersed which can be used to differentiate potential Label predictions better than two variables that are linearly dependent with overlapping points. I also learned that adding a new feature that computers the square root of the sum of the rows featuers squarred adds a number that is more easily identifiable as a normalized attribute to easily differentiate between instances where the row belongs to Label 0 or Label 1. I also learned that bulding a pipeline can be very helpful in cleaning datasets. With pipelines, I don't have to worry about calling multiple functions and returning a value that I will then pass into another function because I can just feed the returned datasets trhough a functional pipeline that will use the return values of the previous function as the input values of the next. This adds efficiency and can potentially make your code look a lot more neat and concise as to what you are doing to clean your dataset before any other operations.
